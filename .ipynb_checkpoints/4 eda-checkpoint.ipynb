```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

> **Images**
- width/ height ratio of each city
- smallest image of each city
- largest image of each city

```python
def get_imgae(folder, city):

    #folder_name = 'origion-img/' + pinole_sold
    images = glob.glob('origion-img/' + folder + '/*.JPG')

    city_id = []
    city_width = []
    city_height = []
    city_size = []
    city_name = []
    city_wh = []

    for image in images:
        original_image = Image.open(image)
        width, height = original_image.size
        image_name = image.replace('origion-img/pinole_sold/','')
        city_id.append(int(image_name[-12:-4]))
        city_width.append(width)
        city_height.append(height)
        city_size.append(width*height)
        city_name.append(city)
        city_wh.append(round(width/height, 1))

        #print(image_name[-12:-4], 'is %4d wide'%(width), 'x %4d high'%(height) ) 
        
    df = pd.DataFrame({'id': city_id, 
                       'width': city_width, 
                       'height': city_height,
                       'size': city_size,
                       'city': city_name,
                       'w/h': city_wh})
    return df
```

```python
elcerrito = get_imgae('elcerrito_sold', 'EL CERRITO')
elsobrante = get_imgae('elsobrante_sold', 'EL SOBRANTE')
sanpablo = get_imgae('sanpablo_sold', 'SAN PABLO')
pinole_like = get_imgae('pinole_sold/like', 'PINOLE')
pinole_unlike = get_imgae('pinole_sold/unlike', 'PINOLE')
hercules = get_imgae('hercules_sold', 'HERCULES')

cities = [elcerrito, elsobrante, sanpablo, pinole_like, pinole_unlike, hercules]
df = pd.concat(cities)
print(df.shape)
df.columns
```

```python
# minimum image of each city
df.groupby(by='city').min()['size']
```

```python
# maximum image of each city
df.groupby(by='city').max()['size']
```

```python
cities = ['EL CERRITO', 'EL SOBRANTE', 'PINOLE', 'SAN PABLO', 'HERCULES']
fig, axes = plt.subplots(ncols=5, figsize=(18,4))
for i, city in enumerate(cities):
    df[df['city'] == city]['w/h'].hist(ax=axes[i])
    axes[i].set_title(city)
    axes[i].axvline(x=1., ls='--', c='orange')
    axes[i].axvline(x=2., ls='--', c='orange')
    #print('%15s'%(folder) + ': aspect_ratio between 1 and 2, ', between12, '/', len(aspect_ratio))
    
    count = 0
    
    for _ in df[df['city'] == city]['w/h']:
        if (_ > 1.) & (_ < 2.):
            count += 1
    print('%15s'%(city) + ': aspect_ratio between 1 and 2, ', count, '/', len(df[df['city'] == city]['w/h']))
```

> **Dataframe**

```python
house = pd.read_csv('./dataset/sold_clean.csv')
house.head()
```

```python
house.info()
```

```python
fig, axes = plt.subplots(ncols=5, figsize=(18,4))
for i, col in enumerate(['bed', 'bath', 'house size', 'lot size', 'price']):
    if i < 2:
        axes[i].plot(house.groupby('city').mean()[col], marker = 'o')
        axes[i].plot(house.groupby('city').max()[col], marker = 'o')
        axes[i].plot(house.groupby('city').min()[col], marker = 'o')
        axes[i].set_xticklabels(['EL CERRITO', 'EL SOBRANTE', 'PINOLE', 'SAN PABLO', 'HERCULES'], rotation=45)
        axes[i].set_title(col)
        axes[i].legend(['mean', 'max', 'min'])
    elif i >= 2:
        axes[i].plot(house.groupby('city').mean()[col]/1000, marker = 'o')
        axes[i].plot(house.groupby('city').max()[col]/1000, marker = 'o')
        axes[i].plot(house.groupby('city').min()[col]/1000, marker = 'o')
        axes[i].set_xticklabels(['EL CERRITO', 'EL SOBRANTE', 'PINOLE', 'SAN PABLO', 'HERCULES'], rotation=45)
        axes[i].set_title(col + ' (in thousands)')
        axes[i].legend(['mean', 'max', 'min']);
        
```

```python
features = ['bed', 'bath', 'house size', 'lot size', 'price']

fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(18,20))
for i, city in enumerate(['EL CERRITO', 'EL SOBRANTE', 'PINOLE', 'SAN PABLO', 'HERCULES']):
    for j, column in enumerate(features):
        #house[house['city'] == city][column].hist(ax=axes[i,j], bins=20)
        if column == 'price':
            axes[i,j].hist(house[house['city'] == city][column]/1000., bins=20)
            axes[i,j].set_title(city + ' - ' + features[j] + ' (in thousand)')
        else:
            house[house['city'] == city][column].hist(ax=axes[i,j], bins=20)        
            axes[i,j].set_title(city + ' - ' + features[j])
```
