### Someone 's code - open, resize, save images

```python
def resize_image(input_image_path,
                 output_image_path,
                 size):
    original_image = Image.open(input_image_path)
    width, height = original_image.size
    print('The original image size is {wide} wide x {height} '
          'high'.format(wide=width, height=height))
 
    resized_image = original_image.resize(size)
    width, height = resized_image.size
    print('The resized image size is {wide} wide x {height} '
          'high'.format(wide=width, height=height))
    resized_image.show()
    resized_image.save(output_image_path)
```

```python
input_image_path = './test-img/40833585.JPG'
output_image_path =  './test-img/40833585_small.JPG'
size = (600, 400)

resize_image(input_image_path, output_image_path, size)
```

### Resize image with scaling (use thumbnail)
- can't receive exactly size of 600 * 400 image


```python
def scale_image(input_image_path,
                output_image_path,
                width=None,
                height=None
                ):
    original_image = Image.open(input_image_path)
    w, h = original_image.size
    print('The original image size is {wide} wide x {height} '
          'high'.format(wide=w, height=h))
 
    if width and height:
        max_size = (width, height)
    elif width:
        max_size = (width, h)
    elif height:
        max_size = (w, height)
    else:
        # No width or height specified
        raise RuntimeError('Width or height required!')
 
    original_image.thumbnail(max_size, Image.ANTIALIAS)
    original_image.save(output_image_path)
 
    scaled_image = Image.open(output_image_path)
    width, height = scaled_image.size
    print('The scaled image size is {wide} wide x {height} '
          'high'.format(wide=width, height=height))
    

    
input_image_path = './test-img/40833585.JPG'
output_image_path =  './test-img/40833585_small_scaling.JPG'
scale_image(input_image_path, output_image_path, width=600, height=400)
                
```

### Examining a specific data range
Sometimes you want to enhance the contrast in your image, or expand the contrast in a particular region while sacrificing the detail in colors that don't vary much, or don't matter. A good tool to find interesting regions is the histogram. To create a histogram of our image data, we use the hist() function.

```python
from PIL import Image
img = Image.open('./test-img/40833585_small.JPG').convert('LA')
img.save('./test-img/40833585_small_greyscale.png')
```

```python
img = mpimg.imread('./test-img/40833585_small_greyscale.png')
```

```python
lum_img = img[:, :, 0]
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18,12))
axes[0,0].hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
axes[0,1].imshow(img)
axes[1,0].imshow(lum_img)
axes[1,1].imshow(lum_img, clim=(0.0, 0.8)) # see histogram x-axis
axes[1,0].set_title('Before')
axes[1,1].set_title('After');
```

### image to array to binary file

```python
import pickle as p

img = Image.open('./test-img/40833585_small.JPG')
r, g, b = img.split()
```

```python
result = np.array([])

print(np.array(r).shape)

r_arr = np.array(r).reshape(240000)
g_arr = np.array(g).reshape(240000)
b_arr = np.array(b).reshape(240000)

image_arr = np.concatenate((r_arr, g_arr, b_arr))
result = np.concatenate((result, image_arr))
result = result.reshape(1,720000)
```

```python
image_arr.shape, result.shape
```

```python
print('start to save files')

file_path = './test-img/data.bin'
with open(file_path, mode='wb') as f:
    p.dump(result, f)

print('files are saved')
```

```python
file_path = './test-img/data.bin'
with open('./test-img/data.bin', mode='rb') as f:
    arr = p.load(f)
```

```python
rows = arr.shape[0]
arr = arr.reshape(rows, 3, 400, 600)


for index in range(rows):
    a = arr[index]
    # get r, g, b back
    r = Image.fromarray(a[0]).convert('L')
    g = Image.fromarray(a[1]).convert('L')
    b = Image.fromarray(a[2]).convert('L')
    image = Image.merge("RGB", (r, g, b))

plt.imshow(image);
```
