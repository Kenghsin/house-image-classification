```python
import numpy as np
import pandas as pd
```

> **Function for data cleaning**

```python
def format_id(x):
    if '-' in x:
        return int(x.replace('.JPG','')[:-2])
    else:
        return int(x.replace('.JPG', '')[-8:])
    
def format_bath(x):
    x_split = x.split()
    if len(x_split) == 4:
        return float(x_split[0]) + 0.5
    elif len(x_split) == 2:
        return float(x_split[0])
    
def format_house_size(x):
    x_split = x.split()
    if x_split[0] == 'N/A':
        return 0
    elif x_split[0] != 'N/A':
        return int(x_split[0].replace(',', ''))
```

```python
def df_cleaning(df, city):
    df['id'] = df['id'].map(format_id)
    df['zip code'] = df['adress'].map(lambda x: int(x[-5:]))
    df['adress'] = df['adress'].map(lambda x: x[:-19])
    df['city'] = city
    df['bed'] = df['bed'].map(lambda x: int(x.replace(' Beds', '')))
    df['bath'] = df['bath'].map(format_bath)
    df['house size'] = df['house size'].map(format_house_size)
    df['lot size'] = df['lot size'].map(lambda x: int(x.replace(', ', '')))
    df['price'] = df['price'].map(lambda x: int(x.split()[1].replace('$','').replace(',', '')))
    df.to_csv('./dataset/'+city.replace(' ','').lower()+'_sold_clean.csv', index=False)
```

> **Final cleaning**
- 10 missing values in **house size**, drop it
- 49 missing values in **lot size**, most of them are condo, make it the same as house size
- save to `sold_clean.csv`

```python
for _ in df.columns:
    print('%2d' % df[df[_] == 0].shape[0], 'zeros in', _)
```

```python
df[df['house size'] == 0]
```

```python
# drop house size = 0
print(df.shape)
df = df[df['house size'] != 0]
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
# set missing value in lot size is equal to value in house size
lot_size  = list(df['lot size'])
house_size = list(df['house size'])
print(len(lot_size), len(house_size))

for i, size in enumerate(lot_size):
    if size == 0:
        lot_size[i] = house_size[i]
         
np.count_nonzero(lot_size) 
```

```python
lot_size_1 = pd.DataFrame(lot_size, columns=['lot_size1'])
df.shape, lot_size_1.shape
```

```python
df = pd.concat([df, lot_size_1], axis =1)
```

```python
df.to_csv('./dataset/sold_clean.csv', index=False)
```
